### prepare data for train

```python
y=pd.get_dummies(df['diagnosis'])
x=df.drop('diagnosis',axis=1)
y=y.drop('M',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

### Feature selection

```python
cor = x_train.corr()
plt.figure(figsize=(30,20))
sns.heatmap(cor, cmap=plt.cm.CMRmap_r,annot=True)
plt.show()  
```

```python
# If the value is near to 1 that means those two features are correlated and we can drop any one of them.
def correlation(dataset, threshold):
    dictionary = {}
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                                colname = corr_matrix.columns[i]                  
                                dictionary[colname]=corr_matrix.columns[j]
    return dictionary 
```

```python
# Getting correlated features

corr_features = correlation(x_train,0.9)

corr_features
```

```python
corr_features=['radius_mean','area_mean','concavity_mean','radius_se','area_se','texture_mean','radius_worst','area_worst']
```

```python
x_train=x_train.drop(corr_features,axis=1)
x_test=x_test.drop(corr_features,axis=1)
```

### classification model

```python
p1=df.groupby('diagnosis')['radius_mean'].count()
p1.plot.barh()
```

#### this may lead to imbalanced classification

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=0)
X_resampled, y_resampled = sm.fit_resample(x_train, y_train)
```

### 1.Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

model=GaussianNB()
model.fit(X_resampled, y_resampled)
prediction_results = model.predict(x_test) 
repo=classification_report(y_test,prediction_results)

print(repo)

cm = confusion_matrix(y_test,prediction_results)
sns.heatmap(cm, cmap=plt.cm.CMRmap_r,annot=True)
plt.show() 
```

### 2.LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

model =LogisticRegression()
model.fit(X_resampled, y_resampled)
prediction_results = model.predict(x_test)
repo = classification_report(y_test, prediction_results)

print(repo)

cm = confusion_matrix(y_test, prediction_results)
sns.heatmap(cm, cmap="CMRmap_r", annot=True)
plt.show()
```

### 3. SVM

```python
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

model=svm.SVC()
model.fit(X_resampled, y_resampled)
prediction_results = model.predict(x_test) 
repo=classification_report(y_test,prediction_results)

print(repo)

cm = confusion_matrix(y_test,prediction_results)
sns.heatmap(cm, cmap=plt.cm.CMRmap_r,annot=True)
plt.show() 
```

### 4.Decision Tree

```python
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

model=tree.DecisionTreeClassifier()
model.fit(X_resampled, y_resampled)
prediction_results = model.predict(x_test) 
repo=classification_report(y_test,prediction_results)

print(repo)

cm = confusion_matrix(y_test,prediction_results)
sns.heatmap(cm, cmap=plt.cm.CMRmap_r,annot=True)
plt.show() 
```
